```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Data dibawah merupakan *return* dan risiko untuk kumpulan saham dari indeks LQ45 hasil optimisasi *multi-objective* dengan menggunakan sebanyak 11 nilai bobot. 

```python
data = pd.DataFrame({'sigma_9' : [0.069282,0.059046,0.042426,0.047958,
                                  0.074833,0.071414,0.054772,0.061644,
                                  0.064807,0.04757,0.05745],
                     'mu_9' : [0.0051,0.0145,0.0116,0.002,
                               0.0253,0.0053,0.0018,0.0126,
                               0.0119,0.0095,0.0074],
                     'sigma_14' : [0.06403,0.05657,0.06164,0.04583,
                                   0.04243,0.05916,0.05745,0.04796,
                                   0.05099,0.05128,0.07013],
                     'mu_14' : [0.0038,0.0032,0.019,0.0056,
                                0.0021,0.008,0.003,0.0022,
                                0.0115,0.004,0.01]})
```

```python
plt.scatter(data['sigma_9'], data['mu_9'], c='red')
plt.xlabel('Sigma')
plt.ylabel('Return')
plt.title('Pareto Front Saham LQ45 Kendala Cardinality dengan K = 9',pad = 20)
plt.annotate("Bobot \u03BB = 0.4 (Profit : Rp12.650.000) ", xy=(0.074833, 0.0253), xytext=(0.074833 - 0.03, 0.0253 - 0.0001), arrowprops=dict(arrowstyle="->"))
plt.annotate("Bobot \u03BB = 0.6 (Profit : Rp900.000) ", xy=(0.054772, 0.0018), xytext=(0.054772 - 0.01, 0.0018 + 0.004), arrowprops=dict(arrowstyle="->"))

plt.savefig('grafik_k9.png')
```

```python
plt.scatter(data['sigma_14'], data['mu_14'], c='green')
plt.xlabel('Sigma')
plt.ylabel('Return')
plt.title('Pareto Front Saham LQ45 Kendala Cardinality dengan K = 14',pad = 20)
plt.annotate("Bobot \u03BB = 0.2 (Profit : Rp9.500.000) ", xy=(0.06164, 0.019), xytext=(0.06164 - 0.008, 0.019 - 0.0025), arrowprops=dict(arrowstyle="->"))
plt.annotate("Bobot \u03BB = 0.4 (Profit : Rp1.050.000) ", xy=(0.04243, 0.0021), xytext=(0.04243 + 0.001, 0.0021 + 0.002), arrowprops=dict(arrowstyle="->"))

plt.savefig('grafik_k14.png')
```
